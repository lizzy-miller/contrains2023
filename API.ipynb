```python
import numpy as np
import pandas as pd
import requests 
import os 
import json
```

```python
congressapi = os.getenv('congressapi')
opensecretsapi = os.getenv('opensecretapi')
```

## Congress Bills API for the 118th Congress 

```python
root = 'https://api.congress.gov/v3'
congress = '118'
endpoint = f'/bill/{congress}'
```

```python
parameters = {'format': 'json',
             'offset': 0,
             'limit': 250,
             'sort': 'updateDate+desc',
             'api_key': congressapi}
```

```python
r = requests.get('https://httpbin.org/user-agent')
useragent = json.loads(r.text)['user-agent']
```

```python
headers = {'User-Agent': useragent,
          'From': 'zrc3hc@virginia.edu'}
```

```python
r = requests.get(root + endpoint, 
                params = parameters, 
                headers = headers)
```

```python
r
```

```python
myjson = json.loads(r.text)
```

```python
myjson['bills'][0]['number']
```

```python
[x['number'] for x in myjson['bills']]
```

```python
with open('bills.json', 'w') as f:
    json.dump(myjson, f)
```

```python
billsdf = pd.json_normalize(myjson, 
                           record_path = ['bills'])
```
