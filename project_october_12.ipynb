```python
states = ['AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME', 'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM', 'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
```

```python
pd.Dataframe = []
```

```python
results = pd.DataFrame()
for i in range(len(states)):
    root = 'http://www.opensecrets.org/api/'
    parameters = {'method': 'getLegislators',
        'apikey': opensecretsapi,
        'id': states[i], 
        'output': 'json'}
    headers = {'User-Agent': useragent,
        'From': 'zrc3hc@virginia.edu'}
    r = requests.get(root, params = parameters, headers=headers)
    myjson = json.loads(r.text)
    info = []
    for x in range(len(myjson['response']['legislator'])):
        dictionary = myjson['response']['legislator'][x]['@attributes']
        info.append(dictionary)
    result = pd.DataFrame.from_dict(info)
    results = pd.concat([results, result])



    
    
```
